```python
# joining the two datasets (swipes and units ) to pull all data in swipes_data with corresponding data in units
```

```python
cobble = pd.merge (Swipes, Units, how = 'left', left_on = 'Swiped card', right_on = 'Unit ID')
```

```python
cobble.head(10)
```

```python
# Creating a value map dictionary for mapping specific values in the 'Is event' column.
valmap= {1.0: 'True', 0.0: 'False'}
# Apply the value map into a new column.
cobble['Is event'] =cobble['Is event'].map(valmap)
```

```python
cobble.head()
```

```python
cobble.info()
```

```python
#number of missing values per column
```

```python
cobble.isnull().sum()
```

```python
#percentage of missing values
```

```python
percent_missing = cobble.isnull().sum() * 100 / len(cobble)
```

```python
percent_missing
```

```python
#filling the null values in 'is event' and 'Price tier' columns  as they are more than 5% of the data set
```

```python
cobble['Is event'].replace(np.nan, 'unknown', inplace= True) #using 'unknown' seem logical
```

```python
cobble.info()
```

```python
#filling the null values in 'Price tier' with '-1'(unknown) as it seemed more logical since 0 is also  a value
```

```python
cobble['Price tier'].replace(np.nan, -1.0, inplace= True)
```

```python
cobble.info()
```

```python
cobble.isnull().sum()
```

```python
#dropping the remaining rows that contain null values as they are less 5% of the dataset 
```

```python
cobble.dropna(axis = 0, how ='any', inplace = True)
```
